```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

tokens_path = "D:\Various-APITokens-Details\spotify.txt"
access_token = open(tokens_path,"r").read().split()
client_id = access_token[2]
client_secret = access_token[-1]
```

```python
auth_manager = SpotifyClientCredentials(client_id= client_id, client_secret=client_secret)

sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
artist_name = "Arijit Singh"#'Black Sabbath'#"ozzy osbourne"
```

# Searching Artist

```python

artists = sp.search(q='artist:'+artist_name,type='artist', limit=10)
```

```python
artist_uri = artists['artists']['items'][0]['uri']
print(artist_uri)
```

```python
results = sp.artist_top_tracks(artist_uri)

for track in results['tracks'][:10]:
    print('track    : ' + track['name'])
    if track['preview_url'] is not None:
        print('audio    : ' + track['preview_url'])
    print('cover art: ' + track['album']['images'][0]['url'])
    print()
```
